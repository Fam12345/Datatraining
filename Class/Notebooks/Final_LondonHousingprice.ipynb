```python
c_properties = pd.melt(properties_t, id_vars= ['London_Borough', 'ID'])
```

```python
### The 32 London Districts 

columnsf = [
        'Barking & Dagenham', 'Barnet', 'Bexley', 'Brent',
        'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield', 'Greenwich',
        'Hackney', 'Hammersmith & Fulham', 'Haringey', 'Harrow', 'Havering',
        'Hillingdon', 'Hounslow', 'Islington', 'Kensington & Chelsea',
        'Kingston upon Thames', 'Lambeth', 'Lewisham', 'Merton', 'Newham',
        'Redbridge', 'Richmond upon Thames', 'Southwark', 'Sutton',
        'Tower Hamlets', 'Waltham Forest', 'Wandsworth', 'Westminster'
    ]



c_properties2=c_properties[c_properties['London_Borough'].isin(columnsf)]
c_properties2



```

```python
c_properties2['London_Borough'].unique()
```

```python
c_properties2.isnull().sum()
```

```python
c_properties2.dtypes
```

```python
c_properties2['Average_price']= pd.to_numeric(c_properties2['Average_price'])
```

```python
c_properties2.dtypes
```

```python
c_properties2
```

```python
c_properties2['Year'] = c_properties2['Month'].apply(lambda t: t.year)
```

```python
c_properties2
```

```python
DF=  c_properties2.groupby(by=['London_Borough', 'Year']).mean()
```

```python
DF
```

```python
DF.index
```

```python
DFF= DF.reset_index()
```

```python
DFF
```

```python
DFF.columns
```

```python
def create_price_ratio(i):
    start = float(i['Average_price'][i['Year']==1998])
    end = float(i['Average_price'][i['Year']==2018])
    ratio = [end/start]
    return ratio

```

```python
Borough_price_ratio= {}
for b in DFF['London_Borough'].unique():
    borough = DFF[DFF['London_Borough'] == b]
    Borough_price_ratio[b] = create_price_ratio(borough)
```

```python
print(Borough_price_ratio)
```

```python
InterestedDataFrame= pd.DataFrame(Borough_price_ratio)
```

```python
print(InterestedDataFrame)
```

```python
InterestedDataFrame_t= InterestedDataFrame.T
```

```python
InterestedDataFrame_t
```

```python
Final_District= InterestedDataFrame_t.reset_index()
```

```python
Final_District
```

```python
Final_District.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
```

```python
Final_District
```

```python
greatest_average_increaseinprice= Final_District.sort_values(by='2018',ascending=False)
```

```python
greatest_average_increaseinprice
```

```python
priceincreasebygraph= greatest_average_increaseinprice[['Borough','2018']].plot(kind='bar')

priceincreasebygraph.set_xticklabels(greatest_average_increaseinprice.Borough)

plt.title('Average increase in housing prices over two decades')
plt.xlabel('London Boroughs')
plt.ylabel('Average Price increase')
```

Conclusion:

Thus, according to the data, the greatest increase in housing price is showen in Hackney district and comparatively the lowest is seen in the Hounslow borough.
